```python
from keras.layers import Input, Dense
from keras.models import Model
from keras.models import load_model
# define the input shape and number of latent dimensions

class Inte_omics:

    def __init__(self,X,latent_dim=2):
        self.X= X
        self.input_shape = self.X.shape[1]
        self.latent_dim = 2


    def train_model(self,X, model_name,epochs=20, batch_size=1):
    # define the encoder layers
        self.model_name = model_name
        inputs = Input(shape=self.input_shape)
        encoded = Dense(256, activation='relu')(inputs)
        encoded = Dense(128, activation='relu')(encoded)
        encoded = Dense(64, activation='relu')(encoded)
        encoded = Dense(32)(encoded)
        encoded = Dense(self.latent_dim)(encoded)

        # define the decoder layers
        decoded = Dense(32, activation='relu')(encoded)
        decoded = Dense(64, activation='relu')(decoded)
        decoded = Dense(128, activation='relu')(decoded)
        decoded = Dense(256)(decoded)
        decoded = Dense(self.input_shape)(decoded)

        # create the autoencoder model
        autoencoder = Model(inputs, decoded)

        # compile and fit the model
        autoencoder.compile(optimizer='adam', loss='mse')
        autoencoder.fit(self.X, self.X, epochs=epochs, batch_size=batch_size)
        autoencoder.save(self.model_name)

        return "Model Saved !!!"

    def get_embedding(self):
        autoencoder = load_model(self.model_name)
        encoder = Model(inputs=autoencoder.input, outputs=autoencoder.layers[5].output)
        middle_layer_output = encoder.predict(self.X)

        return middle_layer_output

    
    
    
```

```python
from keras.layers import Input, Dense
from keras.models import Model
from sklearn.utils import resample
import numpy as np 

# Define the base autoencoder architecture

def train_boosted(X, epochs, boosts, name):
    inputs = Input(shape=(X.shape[1],))
    encoded = Dense(256, activation='relu')(inputs)
    encoded = Dense(128, activation='relu')(encoded)
    encoded = Dense(64, activation='relu')(encoded)
    encoded = Dense(32,activation="sigmoid")(encoded)
    encoded = Dense(8)(encoded)

    decoded = Dense(32, activation='relu')(encoded)
    decoded = Dense(64, activation='relu')(decoded)
    decoded = Dense(128, activation='relu')(decoded)
    decoded = Dense(256)(decoded)
    decoded = Dense(X.shape[1])(decoded)



    autoencoder = Model(inputs , decoded)
    autoencoder.compile(optimizer='adam', loss='mse')

    # Train the base autoencoder on your dataset
    x_train = X
    autoencoder.fit(x_train, x_train,
                    epochs=epochs,
                    batch_size=2,
                    shuffle=True)

    # Define the boosting algorithm
    num_boosting_rounds = boosts
    for i in range(num_boosting_rounds):
        print(i)
        # Compute reconstruction error for each sample
        preds = autoencoder.predict(x_train)
        error = np.mean(np.power(x_train - preds, 2), axis=1)

        # Update the sample weights
        sample_weights = np.ones(x_train.shape[0])
        sample_weights[error > np.percentile(error, 75)] *= 2

        # Resample the training set
        x_train_resampled, _, sample_weights_resampled = resample(x_train, x_train,
                                                                 sample_weights,
                                                                 n_samples=x_train.shape[0],
                                                                 replace=True,
                                                                 random_state=42)

        # Retrain the autoencoder with the updated weights
        autoencoder.fit(x_train_resampled, x_train_resampled,
                        sample_weight=sample_weights_resampled,
                        epochs=50,
                        batch_size=2,
                        shuffle=True)
    autoencoder.save(name)
```
